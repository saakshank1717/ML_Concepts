```python
import pandas as pd
from sklearn import datasets

# Load the iris dataset from sklearn
iris = datasets.load_iris()

# Create a DataFrame with feature names and target names as a 'species' string column
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]

# Map species to numeric labels
label_map = {
    "setosa": 0,
    "versicolor": 1,
    "virginica": 2
}

df["label"] = df["species"].map(label_map)

# Save labeled dataset
df.to_csv("iris_labeled.csv", index=False)

print(df.head())
```

```python
#svc working
from sklearn import svm
clf=svm.SVC()
def plotSVC(title):
  #create a mesh to plot in
  x_min, x_max=X[:, 0].min()-1, X[:, 0].max()+1
  y_min, y_max=X[:, 1].min()-1, X[:, 1].max()+1
  h=(x_max/x_min)/100
  xx,yy=np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
  plt.subplot(1,1,1)
  z=svc.predict(np.c_[xx.ravel(), yy.ravel()])
  z=z.reshape(xx.shape)
  plt.contourf(xx,yy,z, cmap=plt.cm.Paired, alpha=0.8)
  plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
  plt.xlabel('Sepal Length')
  plt.ylabel('Sepal Width')
  plt.xlim(xx.min(),xx.max())
  plt.title(title)
  plt.show()
```

```python
kernels=['linear', 'rbf', 'poly']
for kerneln in kernels:
  # Train SVC on only the first two features for 2D plotting
  svc=svm.SVC(kernel=kerneln).fit(X[:, :2],y)
  plotSVC('kernel=' + str(kerneln))
```

```python
gammas=[0.1,1,10,100]
for gamma in gammas:
  svc=svm.SVC(kernel='rbf', gamma=gamma).fit(X[:, :2],y)
  plotSVC('gamma=' + str(gamma))
```

```python
cs=[0.1,1,10,100,1000]
for c in cs:
  svc=svm.SVC(kernel='linear', C=c).fit(X[:, :2],y)
  plotSVC('C=' + str(c))
```

```python
degrees=[0,1,2,3,4,5,6]
for degree in degrees:
  svc=svm.SVC(kernel='poly', degree=degree).fit(X[:, :2],y)
  plotSVC('degree=' + str(degree))
```
