```python
import pandas as pd
ipl_ball_by_ball = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
ipl = pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
ipl.head()
```

```python
#Inner joining 2 datasets

data = pd.merge(ipl_ball_by_ball, ipl, on='id')
data.head()
```

```python
data.columns
```

```python
#Q1 : Top 5 batsmen in IPL

#Harder Method
batsmen = {}
for index,row in data.iterrows():
    if row['batsman'] in batsmen:
        batsmen[row['batsman']] += row['batsman_runs']
    else:
      batsmen[row['batsman']] = 0

print(batsmen)

top5batsmen = sorted(batsmen.items(), key=lambda item: item[1], reverse=True)[:5]
display(top5batsmen)
```

```python
#Easier method
batsmen = data.groupby('batsman')['batsman_runs'].sum()
top5batsmen = batsmen.nlargest(5)
# top5batsmen = batsmen.sort_values(by='batsman_runs',ascending=False)
top5batsmen
```

```python
#for unique values in a column
data['dismissal_kind'].unique()
```

```python
# Top 5 Bowlers

#hard way
bowlers = data[(data['dismissal_kind'] != 'nan') & (data['dismissal_kind'] != 'run out') & (data['dismissal_kind'] != 'retired hurt') & (data['dismissal_kind'] != 'obstructing the field')]
top5bowlers = bowlers.groupby('bowler')['is_wicket'].sum().nlargest(5)
display(top5bowlers)

#easy way(right way)
bowlers = data[data['dismissal_kind'].isin(['caught', 'bowled', 'lbw','stumped', 'caught and bowled', 'hit wicket'])]
top5bowlers = bowlers.groupby('bowler')['is_wicket'].sum().nlargest(5)
display(top5bowlers)
```

```python
#Orange cap winner in 2008

#hard way(wrong)
#orange_cap = data[data['date'].str.split('/').str[-1] == '2008'].groupby('batsman')['batsman_runs'].sum().nlargest(1)
#display(orange_cap)

#actual way
data['date'] = pd.to_datetime(data['date'],format='mixed')
#format mixed checks if they are in mm/dd/yy or dd/mm/yy
data['Season'] = pd.DatetimeIndex(data['date']).year
#creating new column
orange_cap = data[data['Season'] == 2008].groupby('batsman')['batsman_runs'].sum().nlargest(1)
display(orange_cap)
```

```python
#Purple cap in 2008
bowlers = data[data['dismissal_kind'].isin(['caught', 'bowled', 'lbw','stumped', 'caught and bowled', 'hit wicket']) & (data['Season'] == 2008)]
top5bowlers = bowlers.groupby('bowler')['is_wicket'].sum().nlargest(5)
display(top5bowlers)
```

```python
# Orange cap winner of all seasons
def orange_cap(year):
  orange_cap = data[data['Season'] == year].groupby('batsman')['batsman_runs'].sum().nlargest(1)
  name = orange_cap.index[0]
  runs = int(orange_cap.values[0])
  return name,runs

orange_cap_dict = {}
for year in data['Season'].unique():
  name,runs = orange_cap(year)
  orange_cap_dict[int(year)] = {name:runs}

print(orange_cap_dict)
```

```python
#Purple Cap winner of all seasosn
def purple_cap(year):
  bowlers = data[data['dismissal_kind'].isin(['caught', 'bowled', 'lbw','stumped', 'caught and bowled', 'hit wicket']) & (data['Season'] == 2008)]
  purple_cap = bowlers.groupby('bowler')['is_wicket'].sum().nlargest(1)
  name = purple_cap.index[0]
  wickets = int(purple_cap.values[0])
  return name,wickets

purple_cap_dict = {}
for year in data['Season'].unique():
  name,wickets = purple_cap(year)
  purple_cap_dict[int(year)] = {name:wickets}

print(purple_cap_dict)
```
