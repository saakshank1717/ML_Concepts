```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
df
```

```python
X=df.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN(eps=8,min_samples=4)
```

```python
y=db.fit_predict(X)
```

```python
y
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X[y==0,0],X[y==0,1],c='green',label='cluster1')
plt.scatter(X[y==1,0],X[y==1,1],c='red',label='cluster2')
plt.scatter(X[y==2,0],X[y==2,1],c='blue',label='cluster3')
plt.scatter(X[y==3,0],X[y==3,1],c='yellow',label='cluster4')
plt.scatter(X[y==-1,0],X[y==-1,1],c='black',label='noise')#Noise
plt.show()
)
```
